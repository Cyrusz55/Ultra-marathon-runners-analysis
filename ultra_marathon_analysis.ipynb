```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Ultra-marathon-runners-analysis/TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head()
```

```python
# filter the event distance to either 50miles or 50 kilometers
df[df['Event distance/length'].isin(['50km', '50mi' ])]
```

```python
# filtering further to take those in 2020
df[(df['Event distance/length'].isin(['50km', '50mi' ])) & (df['Year of event'] == 2020)]
```

```python
# filter out only those that happened in USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# now combine all of them
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
# remove (USA) from event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# clean up the athlete age
df2['Athlete year of birth'] = df2['Athlete year of birth'].dropna()
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(2)
```

```python
# drop athlete club, athlete country, athlete year of birth, athlete age category.
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(2)
```

```python
# cleaning up null values
df2.isnull().sum()
```

```python
df2[df2['athlete_age'].isnull() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# check for duplicates
df2[df2.duplicated() == True]
```

```python
# reset the index
df2.reset_index(drop=True)
```

```python
# fix types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# rename the columns
#Year of event	int64
#Event dates	object
#Event name	object
#Event distance/length	object
#Event number of finishers	int64
#Athlete performance	object
#Athlete gender	object
#Athlete average speed	object
#Athlete ID	int64
#athlete_age	float64

df2 = df2.rename(columns={'Year of event': 'year_of_event',
                          'Event dates': 'race_day',
                          'Event name': 'race_name',
                          'Event distance/length': 'race_distance',
                          'Event number of finishers': 'race_number_of_finishers',
                          'Athlete performance': 'athlete_performance',
                          'Athlete gender': 'athlete_gender',
                          'Athlete average speed': 'athlete_average_speed',
                          'Athlete ID': 'athlete_id'})

```

```python
df2.head(2)
```

```python
# reorder columns
df3 = df2[['race_day', 'race_name', 'race_distance', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed', 'year_of_event']]
```

```python
df3.head(2)
```

```python
# find two specific races- Sarasota and Everglades
df3[df3['race_name'] == 'Everglades 50 Mile Ultra Run ']
```

```python
df3[df3['athlete_id'] == 222509]
```

```python
# charts and graphs
import seaborn as sns
sns.histplot(df3['race_distance'])
```

```python
sns.histplot(df3, x = 'race_distance', hue = 'athlete_gender')
```

```python
sns.displot(df3[df3['race_distance'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x = 'race_distance', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart', linewidth = 1)
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender', col = 'race_distance')
```

```python
# Questions to find out from the data
# Difference in speed for 50k, 50mi male to female

df3.groupby(['race_distance', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age groups are the best in the 50m Race (20 + races minimum)
df3.query('race_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count > 19').head(15)
```

```python
# What age groups are the worst in the 50mi race (20 + races min)()
df3.query('race_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count > 11').head(15)
```

```python
# seasons for the data -> Slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
# winter 12 -2

# split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'fall' if x > 8 else 'summer' if x > 5 else 'spring' if x > 2 else 'winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50 Miles only
df3.query('race_distance == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
